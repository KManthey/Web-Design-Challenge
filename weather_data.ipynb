```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

```python
#path for data
weather_csv = 'city_weather.csv'
```

```python
weather_df = pd.read_csv(weather_csv)
weather_df.head()
```

```python
# Get the indices of cities that have humidity over 100%.
max_humidity = weather_df[weather_df["Humidity"] > 100].index
```

```python
# Drop cities with humidity > 100 creating a new DataFrame
clean_weather = weather_df.drop(max_humidity, inplace=False)
clean_weather
```

```python
# Convert the DF to an HTML
clean_weather.to_html('data.html')
```
